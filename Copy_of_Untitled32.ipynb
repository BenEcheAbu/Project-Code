Applying Linear regression

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

def train_linear_regression(X, y):
    #LinearRegression model
    linear_regression = LinearRegression()

    #Splitting the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

    #Fitting the model on the training data
    linear_regression.fit(X_train, y_train)

    return linear_regression, X_train, X_test, y_train, y_test

def evaluate_model(model, X_train, X_test, y_train, y_test):
    #Evaluate model performance on the training set
    linear_regression_train_pred = model.predict(X_train)
    linear_regression_train_r2 = r2_score(y_train, linear_regression_train_pred)

    linear_regression_test_pred = model.predict(X_test)
    linear_regression_test_r2 = r2_score(y_test, linear_regression_test_pred)

    linear_regression_mse = mean_squared_error(y_test, linear_regression_test_pred)
    linear_regression_rmse = np.sqrt(linear_regression_mse)
    linear_regression_mae = mean_absolute_error(y_test, linear_regression_test_pred)

    #Return performance metrics
    return linear_regression_train_r2, linear_regression_test_r2, linear_regression_mse, linear_regression_rmse, linear_regression_mae

#Calling the functions
linear_regression, X_train, X_test, y_train, y_test = train_linear_regression(X, y)
linear_regression_train_r2, linear_regression_test_r2, linear_regression_mse, linear_regression_rmse, linear_regression_mae = evaluate_model(linear_regression, X_train, X_test, y_train, y_test)

#Printting the performance
print("Linear Regression:")
print("Training R-squared:", linear_regression_train_r2)
print("Test R-squared:", linear_regression_test_r2)
print("Mean Squared Error:", linear_regression_mse)
print("Root Mean Squared Error:", linear_regression_rmse)
print("Mean Absolute Error:", linear_regression_mae)
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

def train_random_forest(X, y):
    #Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

    #Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    #RandomForestRegressor
    random_forest = RandomForestRegressor(random_state=5)

    #Fit the model on the training data
    random_forest.fit(X_train_scaled, y_train)

    return random_forest, X_train_scaled, X_test_scaled, y_train, y_test

def evaluate_model(model, X_train, X_test, y_train, y_test):
    #Evaluate model performance on the training set
    random_forest_train_pred = model.predict(X_train)
    random_forest_train_r2 = r2_score(y_train, random_forest_train_pred)

    random_forest_test_pred = model.predict(X_test)
    random_forest_test_r2 = r2_score(y_test, random_forest_test_pred)

    random_forest_mse = mean_squared_error(y_test, random_forest_test_pred)
    random_forest_rmse = np.sqrt(random_forest_mse)
    random_forest_mae = mean_absolute_error(y_test, random_forest_test_pred)

    #Return performance metrics
    return random_forest_train_r2, random_forest_test_r2, random_forest_mse, random_forest_rmse, random_forest_mae

#Calling the functions
random_forest, X_train_scaled, X_test_scaled, y_train, y_test = train_random_forest(X, y)
random_forest_train_r2, random_forest_test_r2, random_forest_mse, random_forest_rmse, random_forest_mae = evaluate_model(random_forest, X_train_scaled, X_test_scaled, y_train, y_test)

#Performing cross-validation
cv_scores = cross_val_score(random_forest, Healthcare_X, y, cv=5, scoring='r2')
cv_r2_mean = np.mean(cv_scores)

#Print the performance metrics
print("Random Forest Regression:")
print("Training R-squared:", random_forest_train_r2)
print("Test R-squared:", random_forest_test_r2)
print("Mean Squared Error:", random_forest_mse)
print("Root Mean Squared Error:", random_forest_rmse)
print("Mean Absolute Error:", random_forest_mae)
print("Cross-Validation R-squared Mean:", cv_r2_mean)
```
